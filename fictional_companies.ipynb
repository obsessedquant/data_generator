# Fictional Companies

These fictional companies are scraped from Fandom.com

```python
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from bs4 import BeautifulSoup as soup

# Instantiate an Options object and add the --headless argument
# This runs the browser in the background
opts = Options()
opts.add_argument(" --headless")

# Set the location of the webdriver
chrome_driver='C:\\Users\\srobi\\Projects\\chromedriver.exe'

# Instantiate a webdriver
driver = webdriver.Chrome(options=opts,executable_path=chrome_driver)

# To scrape a url rather than a local file use this
my_url = 'https://list.fandom.com/wiki/List_of_fictional_companies'
driver.get(my_url)

# Put the page source into a variable and create a BS object from it, parse HTML
page_html = driver.page_source
page_soup = soup(page_html, "html.parser")

# Close the browser
driver.quit()
```

```python
# Tests to see if html has been parsed properly
print('\nTesting html parsing...')
print(page_soup.title.get_text())
print(page_soup.h1.text) # shows header information
print(page_soup.p.text) # shows paragraph information
print(page_soup.body.span.text)
print('Testing completed.\n')
```

```python
# li tags in page_soup gathered into a list and taking the text and moving them into a list

lis = page_soup.find_all("li")

mylist = []
for x in lis:
    mylist.append(x.text)

cleaned_list = mylist[153:552]

# obtain the lists of companies and reference notes for those companies

companies = []
refs = []
for x in cleaned_list:
    try:
        print(x.split(' - ')[1])
        companies.append(x.split(' - ')[0])
        refs.append(x.split(' - ')[1])
    except IndexError as e:
        print("error", e)
```

```python
# Get the industry_categories
# <span class="mw-headline">
industry_categories = page_soup.findAll("span",{"class":"mw-headline"})
print(f'There are {len(industry_categories)} total.') # find total number of containers
first_instance = industry_categories[0].text # look at the first container
print(f'Here is the first title: \n{first_instance}\n')
```

```python
# Get the <span class=new> tags
class_new_spans = page_soup.findAll("span",{"class":"new"})
print(f'There are {len(class_new_spans)} total.') # find total number of containers
class_new_span = class_new_spans[0].text # look at the first container
print(f'Here is the first title: \n{class_new_span}\n')
```

```python
# This section returns a list of all of the items in the <span class="mw-headline">
# Industry categories

industry_list = []

for x in range(len(industry_categories)):
    industry_list.append(industry_categories[x].text)
```

```python
# Removes the items that are not really a part of the industries list
cleaned_industry_list1 = industry_list[1:33]

# Formats industry list to eliminate leading and trailing white spaces and unneeded symbols
import re

cleaned_industry_list2 = []

for x in cleaned_industry_list1:
    y = re.sub(r'[^a-zA-Z0-9\s&,\/]+','',x).strip()
    cleaned_industry_list2.append(y)

cleaned_industry_list2
```
